```python
import smogn
import pandas as pd
```

```python
d = pd.read_csv('../dataset/baseball_train_final4.csv')
d
```

```python
d_smogn_base = pd.DataFrame([], columns = ['NAME', 'PCODE', 'Date', '선발', '타수', '득점', '안타', '2타', '3타', '홈런', '루타',
       '타점', '도루', '도실', '볼넷', '사구', '고4', '삼진', '병살', '희타', '희비', '투구',
       'barrel', '타율', 'LG', 'KIA', 'KT', '키움', '두산', '한화', 'NC', '롯데', '삼성',
       'SSG', '홈경기수', '원정경기수', '장타', '출루', 'OPS'])

for i in range(1,10):
    d_smogn_base = pd.concat([d_smogn_base,smogn.smoter(
    
    ## main arguments
    data =d,           ## pandas dataframe
    y = '출루',          ## string ('header name')
    k = i,                    ## positive integer (k < n)
    samp_method = 'extreme',  ## string ('balance' or 'extreme')

    ## phi relevance arguments
    rel_thres = 0.80,         ## positive real number (0 < R < 1)
    rel_method = 'auto',      ## string ('auto' or 'manual')
    rel_xtrm_type = 'high',   ## string ('low' or 'both' or 'high')
    rel_coef = 2.25           ## positive real number (0 < R)
)])
d_smogn_base = d_smogn_base.drop_duplicates()
d_smogn_base = d_smogn_base[(d_smogn_base.장타 != 0) & (d_smogn_base.출루 != 0)& (d_smogn_base.OPS != 0)]
d_smogn_base = d_smogn_base[d_smogn_base.타수 >= 30]
d_smogn_base
```

```python
d_smogn_longhit = pd.DataFrame([], columns = ['NAME', 'PCODE', 'Date', '선발', '타수', '득점', '안타', '2타', '3타', '홈런', '루타',
       '타점', '도루', '도실', '볼넷', '사구', '고4', '삼진', '병살', '희타', '희비', '투구',
       'barrel', '타율', 'LG', 'KIA', 'KT', '키움', '두산', '한화', 'NC', '롯데', '삼성',
       'SSG', '홈경기수', '원정경기수', '장타', '출루', 'OPS'])

for i in range(1,10):
    d_smogn_longhit = pd.concat([d_smogn_longhit,smogn.smoter(
    
    ## main arguments
    data =d,           ## pandas dataframe
    y = '장타',          ## string ('header name')
    k = i,                    ## positive integer (k < n)
    samp_method = 'extreme',  ## string ('balance' or 'extreme')

    ## phi relevance arguments
    rel_thres = 0.80,         ## positive real number (0 < R < 1)
    rel_method = 'auto',      ## string ('auto' or 'manual')
    rel_xtrm_type = 'high',   ## string ('low' or 'both' or 'high')
    rel_coef = 2.25           ## positive real number (0 < R)
)])
d_smogn_longhit = d_smogn_longhit.drop_duplicates()
d_smogn_longhit = d_smogn_longhit[(d_smogn_longhit.장타 != 0) & (d_smogn_longhit.출루 != 0)& (d_smogn_longhit.OPS != 0)]
d_smogn_longhit = d_smogn_longhit[d_smogn_longhit.타수 >= 30]
d_smogn_longhit
```

```python
d_smogn_base.to_csv('./dataset/baseball_train_base_yj.csv')
d_smogn_longhit.to_csv('./dataset/baseball_train_longhit_yj.csv')
```
