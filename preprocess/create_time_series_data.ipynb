```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('total_data.csv', parse_dates=['DT'])
data.head()
```

```python
query = ('2018-01-01' <= data['DT']) & (data['DT'] <= '2018-12-31')
data2018 = data[query]
data2018
```

```python
new_data_columns = ['주차', '이름', '타순', '선발', '타수', '득점', '안타', '2타', '3타', '홈런', '루타', '타점', '도루', '도실',
                    '볼넷', '사구', '고4', '삼진', '병살', '희타', '희비', '타율', '출루', '장타', 'OPS', '투구',
                    'avLI', 'RE24', 'WPA']
```

```python
temp = pd.DataFrame(columns=new_data_columns)
temp
```

```python
for player in data2018['이름'].unique():
    info_df = data2018[data2018['이름'] == player]
    temp = pd.concat([temp, info_df.groupby(['주차', '이름']).mean().reset_index()], ignore_index=True)
    
temp
```

```python
# 주차별 경기 상대편 정보 추가
team = pd.read_csv('raw_data/TEAM.csv')
team_unique = team['T_NM'].unique()
team_away = ['@' + team for team in team_unique]
team_total = list(team_unique) + team_away
team_total
```

```python
for team in team_total:
    temp[team] = 0
    
temp
```

```python
temp['선발'] = 0
temp
```

```python
for row in data2018.values:
    wknum, name, oppo, seonbal = row[1], row[2], row[3], row[8]
    find = (temp['주차'] == wknum) & (temp['이름'] == name)
    idx = temp[find].index
    temp.loc[idx, '선발'] += seonbal
    temp.loc[idx, oppo] += 1
    
temp
```

```python
temp.to_csv('2018_ts_1week.csv', index=False)
```

```python
new_columns = list(temp.columns) + ['target_주차', 'new_이름', 'target_OPS']
new_columns
```

```python
temp2 = pd.DataFrame(columns=new_columns)
temp2
```

```python
player_df = temp[temp['이름'] == '이진영']
feature_df = player_df[:-1].reset_index(drop=True)
target_df = player_df[['주차', '이름', 'OPS']][1:]
target_df.rename(columns={'주차':'target_주차', '이름':'new_이름', 'OPS':'target_OPS'}, inplace=True)
target_df.reset_index(drop=True, inplace=True)
concat_df = pd.concat([feature_df, target_df], axis=1)
concat_df
```

```python
# 선수별로 한칸씩 내려서 데이터 구성하기

for player in temp['이름'].unique():
    player_df = temp[temp['이름'] == player]
    feature_df = player_df[:-1].reset_index(drop=True)
    target_df = player_df[['주차', '이름', 'OPS']][1:]
    target_df.rename(columns={'주차':'target_주차', '이름':'new_이름', 'OPS':'target_OPS'}, inplace=True)
    target_df.reset_index(drop=True, inplace=True)
    concat_df = pd.concat([feature_df, target_df], axis=1)
    temp2 = pd.concat([temp2, concat_df], ignore_index=True)
    
temp2
```

```python
condition = temp2['target_주차'] - temp2['주차'] > 1
temp2.drop(temp2[condition], inplace=True)
temp2.reset_index(drop=True, inplace=True)
temp2
```
